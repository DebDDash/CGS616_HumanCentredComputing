```python
disaster_metadata = {
    "canada_wildfires_2016": {
        "country": "Canada",
        "region": "Alberta",
        "start": "2016-05-01",
        "end": "2016-05-31"
    },
    "cyclone_idai_2019": {
        "country": "Mozambique",
        "region": "Southeast Africa",
        "start": "2019-03-04",
        "end": "2019-03-21"
    },
    "ecuador_earthquake_2016": {
        "country": "Ecuador",
        "region": "Manabí",
        "start": "2016-04-16",
        "end": "2016-04-30"
    },
    "hurricane_harvey_2017": {
        "country": "USA",
        "region": "Texas",
        "start": "2017-08-17",
        "end": "2017-09-02"
    },
    "hurricane_irma_2017": {
        "country": "USA",
        "region": "Florida / Caribbean",
        "start": "2017-08-30",
        "end": "2017-09-12"
    },
    "hurricane_maria_2017": {
        "country": "Puerto Rico",
        "region": "Caribbean",
        "start": "2017-09-16",
        "end": "2017-10-02"
    },
    "hurricane_matthew_2016": {
        "country": "Haiti",
        "region": "Caribbean",
        "start": "2016-09-28",
        "end": "2016-10-10"
    },
    "italy_earthquake_aug_2016": {
        "country": "Italy",
        "region": "Central Italy",
        "start": "2016-08-24",
        "end": "2016-08-30"
    },
    "kaikoura_earthquake_2016": {
        "country": "New Zealand",
        "region": "Kaikōura",
        "start": "2016-11-14",
        "end": "2016-11-21"
    },
    "puebla_mexico_earthquake_2017": {
        "country": "Mexico",
        "region": "Puebla",
        "start": "2017-09-19",
        "end": "2017-09-26"
    },
    "srilanka_floods_2017": {
        "country": "Sri Lanka",
        "region": "Western Province",
        "start": "2017-05-25",
        "end": "2017-06-02"
    }
}
```

```python
TWITTER_EPOCH = 1288834974657  # milliseconds

def tweet_id_to_datetime(tweet_id):
    timestamp_ms = (int(tweet_id) >> 22) + TWITTER_EPOCH
    return datetime.fromtimestamp(timestamp_ms / 1000, tz=timezone.utc)
```

```python
BASE_DIR = "/Users/debarpita/Desktop/events_set1/data/"  
all_dfs = []

for disaster_folder, meta in disaster_metadata.items():
    folder_path = os.path.join(BASE_DIR, disaster_folder)

    for split in ["train", "dev", "test"]:
        file_path = os.path.join(folder_path, f"{split}.tsv")
        if not os.path.exists(file_path):
            continue

        df = pd.read_csv(file_path, sep="\t")

        # Add split info
        df["split"] = split

        # Disaster metadata
        df["disaster_name"] = disaster_folder
        df["country"] = meta["country"]
        df["location_region"] = meta["region"]

        # Disaster dates
        df["disaster_start_date"] = pd.to_datetime(meta["start"], utc=True)
        df["disaster_end_date"] = pd.to_datetime(meta["end"], utc=True)
        df["year"] = df["disaster_start_date"].dt.year

        # Tweet time
        df["tweet_time_utc"] = df["tweet_id"].apply(tweet_id_to_datetime)

        all_dfs.append(df)
```
